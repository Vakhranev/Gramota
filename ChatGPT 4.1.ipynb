```python
from pathlib import Path
from openai import OpenAI

# –£—Å—Ç–∞–Ω–æ–≤–∏—Ç–µ –≤–∞—à API-–∫–ª—é—á
client = OpenAI(api_key='')

# –°–ø–∏—Å–æ–∫ –º–æ–¥–µ–ª–µ–π –¥–ª—è —Ç–µ—Å—Ç–∞
models = [
    "gpt-4.1-mini",
    "gpt-4.1-nano"
]

# –°–æ–æ–±—â–µ–Ω–∏–µ, –∫–æ—Ç–æ—Ä–æ–µ –±—É–¥–µ—Ç –æ—Ç–ø—Ä–∞–≤–ª–µ–Ω–æ –∫–∞–∂–¥–æ–π –º–æ–¥–µ–ª–∏
messages = [
    {
        "role": "user",
        "content": '''–û—Ç–≤–µ—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å, —É—á–∏—Ç—ã–≤–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é —Å —Å–∞–π—Ç–∞ –ì—Ä–∞–º–æ—Ç–∞.—Ä—É:
                   "–ù—É–∂–Ω–æ –ª–∏ –≤ –¥–∞–Ω–Ω–ª–º —Å–ª—É—á–∞–µ –∑–∞–∫—Ä—ã—Ç—å –ø—Ä–∏—á–∞—Å—Ç–Ω—ã–π –æ–±–æ—Ä–æ—Ç –ø–µ—Ä–µ–¥ ¬´–ò.–ò.–ò–≤–∞–Ω–æ–≤–∞¬ª —Å —É—á–µ—Ç–æ–º —Ç–æ–≥–æ, —á—Ç–æ —Ä–µ—á—å –∏–¥–µ—Ç –æ –Ω–∞–∑–≤–∞–Ω–∏–∏ –º–µ—Ä–æ–ø—Ä–∏—è—Ç–∏—è?

–ù–∞–ø—Ä–∞–≤–ª—è–µ—Ç—Å—è –ø–∏—Å—å–º–æ –æ—Ç –ø—Ä–µ–¥—Å–µ–¥–∞—Ç–µ–ª—è –ö–æ–Ω—Ñ–µ—Ä–µ–Ω—Ü–∏–∏ –ø–æ –≤–æ–ø—Ä–æ—Å—É —Å–æ–∑–¥–∞–Ω–∏—è –Ω–∞ –ë–ª–∏–∂–Ω–µ–º –í–æ—Å—Ç–æ–∫–µ –∑–æ–Ω—ã,
—Å–≤–æ–±–æ–¥–Ω–æ–π –æ—Ç —è–¥–µ—Ä–Ω–æ–≥–æ –∏ –¥—Ä—É–≥–∏—Ö –≤–∏–¥–æ–≤ –æ—Ä—É–∂–∏—è –º–∞—Å—Å–æ–≤–æ–≥–æ —É–Ω–∏—á—Ç–æ–∂–µ–Ω–∏—è –ò.–ò.–ò–≤–∞–Ω–æ–≤–∞."'''
    }
]

# –ó–∞–ø—Ä–æ—Å –∏ –≤—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
for model_id in models:
    print(f"\nüîç –û—Ç–≤–µ—Ç –æ—Ç –º–æ–¥–µ–ª–∏: {model_id}")
    completion = client.chat.completions.create(
        model=model_id,
        messages=messages
    )
    print(completion.choices[0].message.content)
```
